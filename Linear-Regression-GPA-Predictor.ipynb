```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("C:\\Thanusha\\Python\\ML master class\\GPA_data.CSV")
```

```python
data.head()
```

```python
dummy = pd.get_dummies(data[["Gender","Extra_Curricular"]]).astype(int)
```

```python
dummy.head()
```

```python
dummy.drop(["Gender_Female","Extra_Curricular_No"] , axis = 1 , inplace = True)
```

```python
dummy.head()
```

```python
data.drop(["Gender","Extra_Curricular"], axis=1 , inplace = True)
```

```python
data.head()
```

```python
data.drop(["ID"],axis=1 , inplace=True)
```

```python
data.head()
```

```python
data_new = pd.concat([dummy,data], axis=1)
```

```python
data_new.head()
```

```python
x=data_new.iloc [:,:7].values
```

```python
x
```

```python
y=data_new.iloc [:,7].values
```

```python
y
```

```python
x_train,x_test ,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

```python
model= LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_test
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
MSE =mean_squared_error(y_pred, y_test)
MSE
```

```python
RMSE = np.sqrt(MSE)
```
